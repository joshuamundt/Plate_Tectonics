```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import pytz
import numpy as np
import requests
import datetime as dt
import time 
from datetime import datetime
import datetime
from pandas.plotting import register_matplotlib_converters
import scipy.stats as ss
import dateutil.parser
import geoviews as gv
import geoviews.tile_sources as gvts
import holoviews as hv
import geoviews.feature as gf
import cartopy
import cartopy.feature as cf
from geoviews import opts
from cartopy import crs as ccrs
import geopandas as gpd
from shapely.geometry import Point, Polygon
from descartes.patch import PolygonPatch
from matplotlib.collections import PatchCollection
gv.extension('bokeh')

```

```python
#Plate Tectonics data = data
data=pd.read_csv('query.csv',encoding="ISO-8859-1")

```

```python
# 2015 to 2016 data
datass=pd.read_csv('2015.csv',encoding="ISO-8859-1")
```

```python
# divergent
data3=pd.read_csv('divergent.csv',encoding="ISO-8859-1")
```

```python
# transform
data4=pd.read_csv('transform.csv',encoding="ISO-8859-1")
```

```python
latitude=[]
longitude=[]
magnitude=[]
time=[]

for i in range(len(data)):
    latitude.append(data['latitude'][i])
    longitude.append(data['longitude'][i])
    magnitude.append(data['mag'][i])
    time.append(dateutil.parser.isoparse(data['time'][i]))
    
print("the maximum maginute was", np.max(magnitude),"between 2010 and 2020")  
print("the mimmum maginute was", np.min(magnitude),"between 2010 and 2020")
print("the average maginute was", round(np.mean(magnitude),1),"between 2010 and 2020")
```

```python
plt.scatter(time,magnitude)
plt.xticks(rotation=40)
plt.title('Years vs. Earthquake Magnitude')
plt.show()
```

```python
# import background map
street_map = gpd.read_file('Plate_Boundaries__UTIG.shp')
#data.shape
```

```python

# designate coordinate system
crs = {'init':'espc:4326'}
# zip x and y coordinates into single feature
geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
# create GeoPandas dataframe
geo_data = gpd.GeoDataFrame(data,
 crs = crs,
 geometry = geometry)
# adding the "geometry column"
geo_data.head()

```

```python
# create figure and axes, assign to subplot
fig, ax = plt.subplots(figsize=(9,9))
#add.shp mapfile to axes
street_map.plot(ax=ax, alpha=0.4,color='black')
geo_data.plot(column='mag',ax=ax,alpha=9, legend=True,markersize=data['mag']*30)
plt.title('(2010-2020) Location vs. Earthquake Magnitude', fontsize=15,fontweight='bold')
plt.xlim(-135.132,-119.224)
plt.ylim( 36.886,53.020)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()

```

```python
points = gv.Points(data, ['longitude', 'latitude'], ['mag'],
                   label='(2010-2020) Location Vs.Earthquake Magnitude')
#.select(Year=2050).sort('Population').iloc[-20:]
#.select(Longitude=(-170, -50), Latitude=(0, 50))
# MAP OPTIONS =  gvts.OSM , gvts.Wikipedia , gvts.StamenToner , gvts.EsriNatGeo gvts.EsriImagery + gvts.EsriUSATopo + gvts.EsriTerrain + gvts.EsriReference)
(gvts.OSM*
    points.opts(size=gv.dim('mag')*3, fill_color='mag', line_color='grey', cmap='viridis',
    width=500, height=500, global_extent=False, tools=['hover'], show_legend=False)*
    gv.Labels(points).opts(text_font_size='0pt', text_color='red'))
```
